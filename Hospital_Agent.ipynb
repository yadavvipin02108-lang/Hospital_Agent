```python
import pandas as pd

patients = pd.read_csv("patients.csv")
doctors = pd.read_csv("doctors.csv")
appointments = pd.read_csv("appointments.csv")
treatments = pd.read_csv("treatments.csv")
billing = pd.read_csv("billing.csv")
```

```python
print(patients.columns)
print(doctors.columns)
print(appointments.columns)
print(treatments.columns)
print(billing.columns)
```

```python
while True:

    question = input("Ask Question: ").lower()

    if question == "exit":
        break

    elif "doctor" in question:
        print("Total Doctors:", len(doctors))

    elif "patient" in question:
        print("Total Patients:", len(patients))

    elif "appointment" in question:
        print("Total Appointments:", len(appointments))

    elif "bill" in question or "amount" in question:
        print(billing.head())

    else:
        print("Question not understood.")
```
